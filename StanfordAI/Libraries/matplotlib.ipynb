```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([2,8,16,17],[4,16,32,34])
plt.title("First Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Creating a Scatter Plot
height = np.array([23,56,89,75,48,36,52,98,77,55])
width = np.array([12,55,66,23,45,35,78,78,45,44])
plt.xlim(20,100)
plt.ylim(10,80)
plt.scatter(height,width)
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Width")
plt.show
```
